–í –¥–∞—Ç–∞—Å–µ—Ç–µ 16715 —Å—Ç—Ä–æ–∫ –¥–∞–Ω–Ω—ã—Ö –∏ 11 —Å—Ç–æ–ª–±—Ü–æ–≤.

<div class="alert alert-block alert-success">
<b>‚úÖ–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞:</b> –û—Ç–ª–∏—á–Ω–æ, —Å –¥–∞–Ω–Ω—ã–º–∏ –ø–æ–∑–Ω–∞–∫–æ–º–∏–ª–∏—Å—å!  </div>

````python
def pass_value_barh(df):
    count_of_nulls = (df.isnull().sum()).sort_values().round(2)[lambda x: x != 0]
    percent_of_nulls = (df.isnull().sum()/len(df)*100).sort_values().round(2)[lambda x: x != 0]
    
    if len(count_of_nulls) < 2:
        count_of_nulls = (df.isnull().sum()).sort_values().round(2)
        percent_of_nulls = (df.isnull().sum()/len(df)*100).sort_values().round(2)
        
    dict_percent = dict(percent_of_nulls)
    
    ax = (
        (df[percent_of_nulls.index].isna().mean()*100)
        .to_frame()
        .rename(columns = {0:'space'})
        #.query('space > 0')
        .sort_values(by = 'space', ascending = True)
        .plot(kind = 'barh', figsize = (19,6), rot = -5, legend = False, fontsize = 16) 

    )
    ax.set_title('% –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –¥–∞–Ω–Ω—ã—Ö' + "\n", fontsize = 22, color = 'SteelBlue') 
    
    i = -0.1
    for k in dict_percent:
        if dict_percent[k] < 10:
            color = 'green'
            ax.text(dict_percent[k]+0.1, i + 0.09, str(dict_percent[k])+'%', color=color, 
                fontweight='bold', fontsize=12)
        else:
            color = 'red'
            ax.text(dict_percent[k]-2, i + 0.09, str(count_of_nulls[k])+'    '+str(dict_percent[k])+'%', color=color, 
                fontweight='bold', fontsize=12)
        i += 0.98
````

````python
pass_value_barh(data)
````

````python
def graph_analysis(df,column,type_graph,title, name_x='', figsize=(23,10)):
    rot = 90
    if type_graph != 'bar':
        rot = 0
        ax = (
        df.groupby(column)['total_sales'].sum()
        .reset_index()
        .sort_values('total_sales')
        .plot(x=column, y='total_sales', kind=type_graph, 
          legend=None, figsize=figsize, fontsize=20,
            grid=True, rot = rot)
        );
    else:
        ax = (
            df[column].value_counts()
            .reset_index()
            .sort_values('index')
            .plot(x='index', y=column, kind=type_graph, 
              legend=None, figsize=figsize, fontsize=20,
                grid=True, rot = rot)
    );
        ax.set_xlabel(name_x, fontsize = 30)

    ax.set_title(title + "\n", fontsize = 30)
````

````python
#–ø—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –Ω–∞—à–µ–ª –æ—à–∏–±–∫—É –≤ –≥–æ–¥–µ –æ–¥–Ω–æ–π –∏–∑ –∏–≥—Ä —É –∫–æ–Ω—Å–æ–ª–∏ DS
display(data[(data['platform'] == 'DS')&(data['year_of_release'] == 1985)])
data.loc[(data['platform'] == 'DS')&(data['year_of_release'] == 1985), 'year_of_release'] = 2010
````

<div class="alert alert-block alert-success">
<b>‚úÖ–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞:</b> –ö—Ä—É—Ç–æ ,—á—Ç–æ –∑–∞–º–µ—Ç–∏–ª —ç—Ç–æ! </div>

````python
def top_platforms(df, top_x=data['platform'].nunique(), 
                  platform_list=data['platform'].unique().tolist(), 
                  title = "–ì—Ä–∞—Ñ–∏–∫ –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö –ø–æ –≥–æ–¥–∞–º",
                 only_top=False, region='total_sales'):
    top = (
        df.query('platform in @platform_list')
        .groupby(['platform', 'year_of_release'])
        .agg(Sum=(region, 'sum'))
        .reset_index()
        .groupby('platform')['Sum']
        .sum()
        .sort_values(ascending=False)[:top_x]
        .index
        .tolist()
    )
    a = (
        df.query('platform in @top')
        .groupby(['platform', 'year_of_release'])['total_sales']
        .sum()
        .reset_index()
    )
    if only_top == True:
        return top
    else:
        fig = px.line(a, x='year_of_release', y='total_sales', color = "platform",title = title)
        fig.show('png')
````

### –°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –∂–∞–Ω—Ä—ã (—Ç–æ–ø-5)

````python
#–Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ç–æ–ø–∞ –∂–∞–Ω—Ä–æ–≤ –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º —Ä–µ–≥–∏–æ–Ω–µ
def top_genre(df, top_x, region=None):
    
    return (
        df
        .groupby('genre')
        .agg(Sum=(region, 'sum'))
        .reset_index()
        .groupby('genre')['Sum']
        .sum()
        .sort_values(ascending=False)[:top_x]
        .index
        .tolist()
    )
````

````python
fig = make_subplots(rows=1, cols=3,
                    shared_yaxes=True, 
                    subplot_titles=('–°–µ–≤–µ—Ä–Ω–∞—è –ê–º–µ—Ä–∏–∫–∞', '–ï–≤—Ä–æ–ø–∞', '–Ø–ø–æ–Ω–∏—è'))
i = 1

for region in ['na_sales', 'eu_sales', 'jp_sales']:
    top_5 = top_genre(data_actual, 5, region)
    data_filtered = (
                        data_actual
                        .query('genre in @top_5')
                        .groupby(['genre'])
                        .agg(sales=(region, 'sum'))
                        .reset_index()
                        .sort_values(by='sales', ascending=False)
                    )
    data_filtered['sales'] = round(data_filtered['sales'], 2)
    fig.add_trace(go.Bar(
        name=region,
        x=data_filtered['genre'],
        y=data_filtered['sales'],
        text=data_filtered['sales'],
        marker=dict(line=dict(color='#000000', 
                              width=2))), 1, i)
    i = i + 1
i = 0

fig.update_layout(coloraxis=dict(colorscale='Bluered_r'), 
                  showlegend=True,
                  title={
                      'text': '–°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –∂–∞–Ω—Ä—ã –≤ –∫–∞–∂–¥–æ–º —Ä–µ–≥–∏–æ–Ω–µ (—Ç–æ–ø-5)',
                      'y':0.94,
                      'x':0.5,
                      'xanchor': 'center',
                      'yanchor': 'top'
                  })
fig.show('svg')

fig = make_subplots(rows=1, cols=3,
                    shared_yaxes=True,
                    specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]],
                    subplot_titles=('–°–µ–≤–µ—Ä–Ω–∞—è –ê–º–µ—Ä–∏–∫–∞', '–ï–≤—Ä–æ–ø–∞', '–Ø–ø–æ–Ω–∏—è'))

i = 1

for region in ['na_sales', 'eu_sales', 'jp_sales']:
    top_5 = top_genre(data_actual, 5, region)
    data_filtered = (
                        data_actual
                        .query('genre in @top_5')
                        .groupby(['genre'])
                        .agg(sales=(region, 'sum'))
                        .reset_index()
                        .sort_values(by='sales', ascending=False)
                    )
    data_filtered['sales'] = round(data_filtered['sales'], 2)
    fig.add_trace(go.Pie(labels=data_filtered[['genre']],
                             values=data_filtered['sales'],
                            marker=dict(line=dict(color='#000000', 
                                                  width=2))
                            ), 1, i)
    i = i + 1
i = 0

fig.update_layout(coloraxis=dict(colorscale='Bluered_r'), 
                  showlegend=True)
fig.show('svg')
````
